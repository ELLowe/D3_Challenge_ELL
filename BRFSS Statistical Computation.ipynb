```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
BRFSS_2014 = "data/data.csv"

BRFSS_2014_df = pd.read_csv(BRFSS_2014)
```

```python
BRFSS_2014_df
```

```python
BRFSS_2014_df.columns
```

```python
BRFSS_2014_df = BRFSS_2014_df.drop(['id','state','abbr','povertyMoe','ageMoe','incomeMoe','healthcareLow', 'healthcareHigh','obesityLow', 'obesityHigh','smokesLow',
       'smokesHigh', '-0.385218228'], 1)
```

```python
BRFSS_2014_df.head()
```

```python
BRFSS_2014_df = BRFSS_2014_df.astype('float64')
BRFSS_2014_pearson_df = BRFSS_2014_df.corr(method='pearson', min_periods=1)
BRFSS_2014_pearson_df
```

```python
# importing seaborn for heatmapping:
import seaborn as sns
```

```python
# BRFSS_2014 correlations with a heatmap to highlight greatest correlations:
plt.figure()
sns.heatmap(BRFSS_2014_pearson_df, annot=True,cmap=plt.cm.Reds)
fig = plt.gcf()
fig.set_size_inches(10, 10)

# Save the figure
plt.savefig("BRFSS_2014_pearson_results.png")
plt.show()
```

```python
import scipy.stats as stats
```

```python
# Assuming equal population variance
stats.ttest_ind(BRFSS_2014_df['poverty'], BRFSS_2014_df["age"], equal_var=True)
```

The result of the t-test comparing poverty and age is a p-value of 3.093096525958947e-65. This is odd, because the pearson coeff. was -0.077 I thought the ideal pearson is below 0.05?
